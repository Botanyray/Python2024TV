<h1>Integrals and derivatives<h1>

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

dydx = np.gradient(y,x)
plt.plot(x,y)
plt.show()
plt.figure()
plt.plot(x,dydx)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
dydx = np.gradient(y,x)

y_integral = np.trapz(y,x)
y_integral
```

```python
plt.plot(x,y)
plt.plot(x,dydx)
```

## Plotting 2D functions



```python
x = np.linspace(0, 10, 1000)
y = np.linspace(0, 5, 500)
print(len(x))
print(len(y))
```

```python

xv, yv=np.meshgrid(x,y)
```

```python
len(yv)
```

```python
zv=2*xv**2+2*yv**2
plt.contourf(xv,yv,zv, levels=100)
plt.colorbar()
plt.show()
```

```python
x=np.linspace(-50, 50, 100)
y=np.linspace(-50, 50,200)
xv,yv=np.meshgrid(x,y)
zv=3*xv**2+3*yv**2
#plt.contourf(xv,yv,zv, levels=100)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(xv,yv,zv, 500)
ax.view_init(120,0)
```

# Basic Linear Algebra

## Matrix Operations

```python
A = np.array([[3, 2, 1],[5,-5,4],[6,0,1]])
b1 = np.array([1,2,3])
b2 = np.array([-1,2,-5])
```

```python
A
```

```python
b1
```

```python
#Moltiplicazione matrice 3x3 per vettore 3x1. Il risultato è un vettore 3x1.
A@b1
```

```python
#Prodotto scalare
np.dot(b1, b2)
```

```python
#Prodotto vettore
np.cross(b1, b2)
```

## Systems of Equations

$ 3x + 2y + z = 4$

$ 5x -5y + 4z = 3$

$ 6x +z = 0$

```python
A = np.array([[3, 2, 1],[5,-5,4],[6,0,1]])
c = np.array([4,3,0])
```

```python
np.linalg.solve(A,c)
```

## Finding Eigenvalues

```python
A = np.array([[4,2,2],[2,4,2],[2,2,4]])
```

```python
np.linalg.eig(A)
```

```python
w, v = np.linalg.eig(A)
# printing eigen values
print("Printing the Eigen values of the given square array:\n",w)

# printing eigen vectors
print("Printing Right eigenvectors of the given square array:\n",v[0], 'and', v[1], 'and ', v[2])
```

```python
print(v[:,0])
A@v[:,0]
```

```python
w[0]*v[:,0]
```
