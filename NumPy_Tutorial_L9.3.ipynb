## Question 4:

The power emitted by a blackbody is $P = A \sigma \epsilon T^4$. After measuring the temperature of a star you find that $T(t) = T_0 \frac{1}{1+e^{-kt}}$. Plot the total energy emitted by the star as a function of time using the fact that $E(t) = \int_{t'=0}^{t'=t} P(t') dt' $

```python
kt = np.linspace(0, 3, 100)
P = (1/(1+np.exp(-kt)))**4
E=np.trapz(P,kt)
E=[]
for ii in kt:
    xx=kt[kt<ii]
    P = (1/(1+np.exp(-xx)))**4
    E.append(np.trapz(P,xx))
      
plt.plot(kt,E)
```

```python
plt.plot(kt,E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
plt.show()
```

## Question 3 (tougher)

An electric field is given by $\vec{E}(z,t) = E_0 \cos (z-t) \hat{x} + 2E_0 \cos(z-t+\pi/2) \hat{y}$. 

1. Find the magnetic field for all $z \in [0,4\pi]$ and $t \in [0,10]$ using $c\vec{B}(z,t) = \hat{z} \times \vec{E}(z,t)$
2. Compute the Poynting vector $\vec{S} = \vec{E} \times \vec{B}$ for all $z$ and $t$

```python
z = np.linspace(0, 4*np.pi, 100)
t = np.linspace(0, 10, 100)

tv, zv = np.meshgrid(t, z)

Ex = np.cos(zv-tv)
Ey = 2*np.cos(zv-tv+np.pi/2)
Ez = 0*zv
plt.contourf(tv,zv,Ex, levels=100)
```

$E_x$ as function of $t$ at $z=0$

```python
plt.plot(t, Ex[0])
plt.plot(t, Ey[0])
```

$E_x$ as function of $z$ at $t=0$

```python
plt.plot(Ex[:,0])
```

```python
Bx=np.copy(Ex)
By=np.copy(Ex)
Bz=np.copy(Ex)
r,c=np.shape(Ex)
for zz in range(r):
    for tt in range(c):
        EE=np.array([Ex[zz,tt],Ey[zz,tt],Ez[zz,tt]])
        [BBx, BBy, BBz]= np.cross(np.array([0,0,1]), EE)
        Bx[zz,tt]=BBx
        By[zz,tt]=BBy
        Bz[zz,tt]=BBz
        
```

$E_y$ and $B_x$  as function of $t$ at $z=0$

```python
t = np.linspace(0, 10, 100)
plt.plot(t,Ey[0,:])
plt.plot(t,Bx[0,:])
```

```python
Sx=np.copy(Ex)
Sy=np.copy(Ex)
Sz=np.copy(Ex)
r,c=np.shape(Ex)
for zz in range(r):
    for tt in range(c):
        EE=np.array([Ex[zz,tt],Ey[zz,tt],Ez[zz,tt]])
        BB=np.array([Bx[zz,tt],By[zz,tt],Bz[zz,tt]])
        [SSx, SSy, SSz]= np.cross( EE,BB)
        Sx[zz,tt]=SSx
        Sy[zz,tt]=SSy
        Sz[zz,tt]=SSz
```
